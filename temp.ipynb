```python
import pandas as pd
```

```python
df = pd.read_csv("src/data/iranAdministrativeDivision.csv")
df.head()
```

```python
records = []
for i in range(0 , len(df)):
    record = {
        "name" : df["title"][i],
        "type" : df["type"][i]
    }
    records.append(record)
```

```python
# write {"records" : records} into iranAdministrativeDivision.json
import json
with open("iranAdministrativeDivision.json" , "w") as f:
    f.write(json.dumps({"records" : records} , ensure_ascii= False , indent= 4))
```
